```python
import cv2
import os
import scipy.misc 
import numpy as np
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
img_file = 'E:\\SigTuple_Hack\\Test_Data\\Mask\\' + '5274941329CC-mask.jpg'
```

```python
import cv2
import numpy as np

kernel = np.ones((5,5),np.uint8)

img = cv2.imread(img_file,0)
erosion = cv2.erode(img,kernel,iterations = 1)
opening = cv2.morphologyEx(erosion, cv2.MORPH_OPEN, kernel)
dilation = cv2.dilate(opening,kernel,iterations = 4)
gradient = cv2.morphologyEx(dilation, cv2.MORPH_GRADIENT, kernel)
```

```python
titles = ['Original Image', 'Erosion',
            'Opening','Dilation', 'Gradient' ]
images = [img, erosion,  opening, dilation, gradient]

for i in xrange(5):
    plt.subplot(2,3,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

```python
import os
import scipy.misc 
ip_files = []
train_files = os.listdir('E:\\SigTuple_Hack\\Test_Data\\Mask')
ip_files += [f for f in train_files if f.endswith('.jpg') ]
#ip_files

for _file in ip_files:
    img_file = 'E:\\SigTuple_Hack\\Test_Data\\Mask\\' + _file
    mask_file = 'E:\\SigTuple_Hack\\Test_Data\\Mask\\' + _file[:-9] + '-mask2.jpg'
    

    img = cv2.imread(img_file,0)
    erosion = cv2.erode(img,kernel,iterations = 1)
    opening = cv2.morphologyEx(erosion, cv2.MORPH_OPEN, kernel)
    dilation = cv2.dilate(opening,kernel,iterations = 4)
    
    # save the mask
    scipy.misc.imsave(mask_file, dilation)
```
